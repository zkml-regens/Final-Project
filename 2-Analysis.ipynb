```python
imageDir = "/content/gdrive/MyDrive/ZKML Regens/ml_ready_plants"
annotations_file = "/content/gdrive/MyDrive/ZKML Regens/ml_ready_plants_metadata/annotations_images.csv"
```

```python
annotationsDF = pd.read_csv(annotations_file)
annotationsDF.head()
```

```python
directories = [d for d in os.listdir(imageDir) if os.path.isdir(os.path.join(imageDir, d))]
directories.sort()
print(directories)
```

```python
# What is the target column?

# What is the link between the images in the dataset and the rows in the csvFile.
# Looks like splitting on "_" and taking the first element maps to the genus column (in most cases).
# What maps to the actual image?

#unique_classes = annotationsDF[['class', 'order', 'family', 'genus']].drop_duplicates().values.tolist()
unique_family = annotationsDF[['genus', 'family']].drop_duplicates().values.tolist()

family = {}
for f in unique_family:
  family[f[0]] = f[1]

print(family)

```

```python
unique_class = annotationsDF[['genus', 'class']].drop_duplicates().values.tolist()

gclass = {}
for c in unique_class:
  gclass[c[0]] = c[1]

print(gclass)
```

```python
unique_order = annotationsDF[['genus', 'order']].drop_duplicates().values.tolist()

order = {}
for o in unique_order:
  order[o[0]] = o[1]

print(order)
```

```python
# prompt: get unique genus

unique_genus = annotationsDF['genus'].unique().tolist()
```

```python
from typing import OrderedDict
genus_images_count = {}
genus_family_count = {}
genus_class_count = {}
genus_order_count = {}

def list_files_in_directories(root_dir):
  """Lists files in all directories within a given root directory."""
  for dirpath, dirnames, filenames in os.walk(root_dir):
    if dirpath == imageDir:
      continue

    directory = dirpath.split("/")[-1]
    genus_dir = directory.split("_")[0].split("(")[-1]
    if genus_dir not in unique_genus:
      continue

    if genus_dir in genus_images_count:
      genus_images_count[genus_dir] += len(filenames)
    else:
      genus_images_count[genus_dir] = len(filenames)

  for g in genus_images_count:
    f = family[g]
    c = gclass[g]
    o = order[g]

    if f in genus_family_count:
      genus_family_count[f] += genus_images_count[g]
    else:
      genus_family_count[f] = genus_images_count[g]

    if c in genus_class_count:
      genus_class_count[c] += genus_images_count[g]
    else:
      genus_class_count[c] = genus_images_count[g]

    if o in genus_order_count:
      genus_order_count[o] += genus_images_count[g]
    else:
      genus_order_count[o] = genus_images_count[g]

  return genus_family_count, genus_class_count, genus_order_count

list_files_in_directories(imageDir)
print(genus_family_count)
print(genus_class_count)
print(genus_order_count)
```
